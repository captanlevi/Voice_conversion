```python
import os
import re
```

```python

def rename_spks(main_dir):
    for spk_dir in os.listdir(main_dir):
        if(spk_dir.startswith(".")):
            continue
        spk_dir_path = os.path.join(main_dir,spk_dir)
        for wav in os.listdir(spk_dir_path):
            new_path = os.path.join(spk_dir_path,wav.lower())
            old_path = os.path.join(spk_dir_path,wav)
            os.rename(old_path,new_path)


```

```python
def parseScript(file_path):
    regex = re.compile(r"[0-9]*-[0-9]*-[0-9]*")
    def splitLine(line):
        spk_id = re.match(regex,line)[0]
        content = line[len(spk_id):].strip()
        return spk_id,content.lower()
    
    info = dict()
    with open(file_path, "r") as f:
        for line in f:
            line = line.replace("\n", "").replace("\ufeff", "").strip()
           
            line_id,content = splitLine(line)
            assert line_id not in info
            info[line_id] = content
    return info



        
    
```

```python
def metaCrawler(source,info_dct):

    if(os.path.isdir(source)):
        for item in os.listdir(source):
            metaCrawler(os.path.join(source,item),info_dct= info_dct)
    
    elif(source.endswith(".txt")):
        info_dct.update(parseScript(source))
        
```

```python
meta_dct = dict()
metaCrawler("LibriSpeech/test-clean", meta_dct)
```

```python
def wavCrawler(source,info_dct):
    if(os.path.isdir(source)):
        for item in os.listdir(source):
            wavCrawler(os.path.join(source,item),info_dct)
    
    elif(source.endswith(".flac")):
        write_path,wav = os.path.split(source)
        wav = wav.split(".flac")[0]
        assert wav in info_dct

        write_path = os.path.join(write_path,wav + ".txt")
        with open(write_path, "w") as f:
            f.write(info_dct[wav])



```

```python
wavCrawler("LibriSpeech/test-clean", meta_dct)
```
